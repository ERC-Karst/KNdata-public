```python
# Load idealized graphs    

import pandas as pd
import networkx as nx


G = nx.Graph()


cavename = 'AnastomoticPalmer'
path_to_csv_edges = f'../data/000_Idealized/clean_graph_csv/{cavename}_edges.csv'
path_to_csv_node_pos = f'../data/000_Idealized/clean_graph_csv/{cavename}_node_pos.csv'

#add edges
#----------
#load edge as dataframe
df = pd.read_csv(path_to_csv_edges, delimiter=';')
#create the graph with the edges
G.add_edges_from(zip(df.from_id,df.to_id))

#add node position
#-----------------
#load node position as dataframe. 
df = pd.read_csv(path_to_csv_node_pos, delimiter=';')
#regroup the position information into a dictionnary with x,y as a list (note that the idealized graph only have x,y position)
dict_pos2d = dict(zip(df.id,zip(df.x,df.y)))   
#add the dictionnary of position ot the graph as an attribute named 'pos2d'
nx.set_node_attributes(G,dict_pos2d,'pos2d')  


#plot the graph
#--------------
nx.draw(G, pos=nx.get_node_attributes(G,'pos2d'))
```
